```python
from datetime import datetime
def convert_hour_to_time(hour):
    hour_str = str(hour).zfill(4)  # Convert to string and pad with zeros (e.g., '100' -> '0100')
    return datetime.strptime(hour_str, '%H%M').time()  
```

```python
merged_df4['hour'] = merged_df4['hour'].apply(convert_hour_to_time)
merged_df4['departure_time'] = pd.to_datetime(merged_df4['departure_date']).dt.time
merged_df4['arrival_time'] = pd.to_datetime(merged_df4['estimated_arrival']).dt.time
```

```python
print(merged_df4[['hour', 'departure_time', 'arrival_time']].head())
```

```python
merged_df4  = merged_df4[
    (merged_df4['hour'] >= merged_df4['departure_time']) &
    (merged_df4['hour'] <= merged_df4['arrival_time'])
]
```
